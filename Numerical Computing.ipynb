# Modified Euler

```python
import numpy as np
import math
import pandas as pd


eq = input("Enter derivated equation in terms of 'y' and 't': ")
equation = "lambda t,y: " + eq
f = eval(equation)



eq = input("Enter actual solution y(t) in terms of 't': ")
equation = "lambda t: " + eq
g = eval(equation)



t0 = float(input("Enter lower limit (t0): "))
b = float(input("Enter upper limit (b): "))
y0 = float(input("Enter y0: "))

h = float(input("Enter h: "))

eulerList = []
tList = []
ytList = []
errorList = []

tList.append(t0)
eulerList.append(y0)
errorList.append(0)
ytList.append(g(t0))

while t0<b:

  k1 = float(h*f(t0,y0))
  k2 = float(h*f(t0+h,y0+k1))

  y0 = y0 + (1/2)*(k1+k2)

  eulerList.append(y0)
  
  t0=t0+h
  tList.append(t0)
  yt = g(t0)
  ytList.append(yt)
  errorList.append(yt-y0)


df = pd.DataFrame({
    "ti":tList,
    "Modified Euler":eulerList,
    "y(ti)":ytList,
    "Error":errorList
})

df
```

## ***Mid point Method***

```python
import numpy as np
import math
import pandas as pd


eq = input("Enter derivated equation in terms of 'y' and 't': ")
equation = "lambda t,y: " + eq
f = eval(equation)



eq = input("Enter actual solution y(t) in terms of 't': ")
equation = "lambda t: " + eq
g = eval(equation)



t0 = float(input("Enter lower limit (t0): "))
b = float(input("Enter upper limit (b): "))
y0 = float(input("Enter y0: "))

h = float(input("Enter h: "))

eulerList = []
tList = []
ytList = []
errorList = []

tList.append(t0)
eulerList.append(y0)
errorList.append(0)
ytList.append(g(t0))

while t0<b:

  
  k1 = float(f(t0,y0))
  k2 = float(f(t0+(h/2),y0+(h/2)*k1))
  
  y0 = y0 + h*k2


  eulerList.append(y0)

  t0=t0+h
  tList.append(t0)
  yt = g(t0)
  ytList.append(yt)
  errorList.append(yt-y0)


df = pd.DataFrame({
    "ti":tList,
    "Mid Point":eulerList,
    "y(ti)":ytList,
    "Error":errorList
})

df
```
